2. Explain the difference between mutable and immutable data types with examples?
 * In Python, data types are classified as mutable or immutable based on whether their value can be changed after they are created.

 a. Mutable data types:
 Definition: Mutable data types allow modifications to their contents without changing their identity.
 Ex: Lists, dictionaries, sets, byte arrays.

 b. Immutable Data types:
 Definition: Immutable data types do not allow modification of their contents. Any operation that alters the value will create a new object in memory.
 Ex: Strings, tuples, integers, floats, booleans.

5. Why might you use a set instead of a list in Python?

6. What is a string in Python, and how is it different from a list?
 * A string in Python is an immutable sequence of characters, meaning once it is created, its contents cannot be changed, and it is used primarily for handling and manipulating text, with operations like slicing, concatenation, and various string-specific methods (e.g: .upper(), .replace()). In contrast, a list is a mutable sequence that can store elements of different data types (such as integers, strings, or other lists), and its contents can be modified, including adding, removing, or changing elements after its creation, making it more flexible for managing collections of items. While both strings and lists are sequences.



7. How do tuples ensure data integrity in Python?
  * Tuples ensure data integrity in Python through immutability, meaning once a tuple is created, its elements cannot be altered, added, or removed. This characteristic makes tuples ideal for situations where data consistency is critical. Tuples are more memory-efficient than lists, which is advantageous for storing large datasets where integrity and immutability are critical. Tuples signal to developers that the data is fixed and should not be modified, making code more predictable and easier to debug.



8. What is a hash table, and how does it relate to dictionaries in Python?
 * A hash table is a data structure that stores key-value pairs and allows for fast data retrieval using a process called hashing. In Python, dictionaries are implemented using hash tables, making them highly efficient for operations like lookups, insertions, and deletions. A hash function maps keys to hash values (integers), which determine the index (or bucket) where the key-value pair will be stored in the hash table.
